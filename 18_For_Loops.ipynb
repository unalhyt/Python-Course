```python
# For Loops
# a way of repeating of a certain statement some number of times

for i in range(5):
    print(i) # repeats this function 5 times
```

```python
for j in range(4,10):
  print(j,j**2) # you can use j as a normal variable
```

```python
# soft coding vs hard coding

# 100s of linse of code......

# and then you have a for loop in here

for i in range(100):
  pass # does nothing


# if you put 100 directly inside the loop is hard coding. We manually specify parameters. It can cause lots of mistakes
# if it is related a parameter before and you ut instead of 100 it will be soft coding
# avoid hard coding


# in soft coding

# first identify parameters like:

nIterations = 100

# 100s of linse of code......

for i in range(nIterations):
  pass
```

```python
# For-loop and if-else statements

for n in range(18):
  if n%2 == 0:
    print(n, 'is even number')
  else:
    print(n, 'is odd number')
```
