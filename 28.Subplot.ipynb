```python
import matplotlib.pyplot as plt
import numpy as np


plt.subplot(1, 2, 1) # First two numbers are about the geometry. 1 row 2 columns. 1 means first subplot
plt.plot(np.random.randn(10))


plt.subplot(1, 2, 2) #
plt.plot(np.random.randn(10))

plt.show()
```

```python
for i in range(9):
  plt.subplot(3, 3, i+1)
  plt.plot(np.random.randn(10))

plt.show()

# this presentation is better for simple cases
```

```python
fig, ax = plt.subplots(1,3,figsize=(10,3)) # figsize defines the size of the plots
# ax # this gives an array

x = np.arange(10)

ax[0].plot(x,x**2,'b') # ax is an array so you can access individual element of it.
ax[1].plot(x,np.sqrt(x),'r')
ax[2].plot(x,x**3,'g')
plt.show()
```

```python
fig,ax = plt.subplots(2,2)
print(np.shape(ax))

# ax[0].plot(np.random.randn(4,4)) # a[0] you cannot access like this
ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))
plt.show()
```
