```python
import numpy as np
n = 2 # for soft coding use this parameter
r = np.zeros(n) # it is very common to initialize variables with zeros

for i in range(n):
  r[i] = i**2

print(r)
# r is a numpy array
```

```python
#Exercise: time all three methods
# 1) list append
# 2) numpy append
# 3) numpy initialization
N = 10000

import time

t = [0]*3

# list append
tic = time.perf_counter()
r = []
for i in range(N):
  r.append(i**2)
t[0] = (time.perf_counter()-tic)*1000 # in miliseconds
print()
# numpy append
tic = time.perf_counter()
r = np.array([])
for i in range(N):
  r = np.append(r,i**2)
t[1] = (time.perf_counter()-tic)*1000

# numpy initialization
tic = time.perf_counter()
r = np.zeros(N)
for i in range(N):
  r[i] = i**2
t[2] = (time.perf_counter()-tic)*1000

print(t)


```
