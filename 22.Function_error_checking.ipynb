```python
# Let's create a function that will input 2 numbers and output their product

def funfun(x,y):
  z = x*y
  return z
```

```python
funfun(2,3)
```

```python
funfun(2,'hi')  # The function should present an error message if one of the inputs are not numbers
```

```python
def funfun2(x,y):
  # check that inputs are numeric
  if not isinstance(x,int):
    raise Exception('input x must be a number!')
  z = x*y
  return z
```

```python
funfun2('hi',2)
```

```python
funfun2(2.5,2) # error because of x must be an integer
```

```python
def funfun3(x,y):
  # check that inputs are numeric
  if not isinstance(x,(int,float)):
    raise Exception('input x must be a number!')

  z = x*y
  return z
```

```python
funfun3(2.5,2)
```

```python
def funfun4(x,y):
  # check that x is numeric
  if not isinstance(x,(int,float)):

    # check if x can be converted into a number
    if isinstance(x,str) and x.isnumeric():
      x = float(x)
      print('Note: converted x to float')
    else:
      raise Exception('input x must be a number!')

  z = x*y
  return z
  # you can copy and paste same if statments for checking y input
```

```python
funfun4('2',3)
```

```python
funfun4('asd2',3)
```
