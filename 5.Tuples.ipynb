```python
# Tuples
# Similar to list
# For Tuple, instead of [] we use ()
# () is also used for functions.

atuple = (3, '4',3)
alist  = [3, '4',3]
#atuple
#atuple*3
atuple.count(3)


# tuples are immutable means not possible to change!!!
# Lists are mutable

print(atuple)
print(alist)

alist[0] = 5
# atuple[0] = 5 it gives error. it is not possible to change something inside the tuple

atuple = (5, '4', 3) # you can change entire tuple to something else

print(atuple)
print(alist)
```

```python
# converting a tuple into a list

print(type(atuple))
print(type(list(atuple)))

v = list(atuple)

print(v)

v[0] = 254456
print(v)
```

```python
# why do we need tuple?
# because we can use it as a storage for a data that will be never changed.
# data inside the tuple will be locked and never changed
```
