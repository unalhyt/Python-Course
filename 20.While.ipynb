```python
# while loops
# while works with a boolean conditional
# as long as conditional is True, it continues
# when conditional becomes False, while loop stops

toggle = True
i = 0

while toggle:
  print(i)
  i += 1
  if i==8:
    toggle = False
```

```python
toggle = True
i = 0

while toggle:
  print(i)
  i += 1
  if i==8:
    toggle = False
# runtime interrupt to exit
```

```python
# another way to exit a while loop
i = 0
while True:
  print(i)
  i += 1
  if i==8:
    break
```

```python
# another way to exit a while loop
i = 0
while i<8:
  print(i)
  i += 1
```

```python
# for-loop vs while loop
# for-loop: prefer when you know in advance the number of iterations
# while-loop: prefer when you don't know in advance the number of iterations
# operations like optimization, you can not know the number o iterations
```

```python
# Exercise: make a Poisson-like counter
# parameter lambda. positive number(10)
# start at 1, multiply by random number with uniform distributed number([0,1])
# target number is smaller than exp(-lambda)
# count the number of iterations until target<exp(-lambda)
```

```python
import numpy as np

def poissonCounter(lam):
  # initialize some variables
  counter = 0
  currval = 1
  target = np.exp(-lam)
  # run the algorithm(multiplication)
  while currval >= target:
    currval *= np.random.uniform(0,1)
    counter += 1
  # return the tesult(counter)
  return counter

poissonCounter(10)
```

```python
p = np.zeros(100)
for i in range(100):
  p[i] = poissonCounter(10)

print(p)
```

```python
# same functionality but harder to read
def poissonCounter(l):
  k,p,t = 0,1,np.exp(-l)
  while p>t:
    p *= np.random.uniform(0,1)
    k += 1
  return k

poissonCounter(10)
```
