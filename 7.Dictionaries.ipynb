```python
# Dictonary
# it is very similar to structures in MATLAB
D = dict() # initializing a dictionary called variable D
D
```

It will be helpful to understand the difference between parantheses in python


*   () is used for functions and tuples
*   [] is used for indexing and slicing and lists
*   {} is used for dictionaries



```python
# The way to create a dictionary is specifying a key/value pair
# Key is like a word in a dictionary
# The value is the definitiona of that word.

D['name'] = 'Mike'
D['AgeRange'] = [20,50]

D
```

*   Key and value are linked by :(column sign)
*   Different key/value pairs are seperated by , (comma sign)

```python
# A dictionary can be created by a different way
# You can enter key/value pairs directly into a curly brackets {} to create a Dictionary

E = {'name': 'Mike', 'AgeRange': [20, 50]}
E
```

```python
# To extract(retrieve) a value from a Dictionary: Call a key from the dictionary
# This is the way we assign a value a key

D['name']
```

```python
# Normally, there will be more than two keys in a Dictionary.
# To know, what are the available keys in the dictionary, we will use a method:

D.keys()
```

```python
# To check a key is inside the Dictionary:
# Key names are case sensitive. That means that 'name' and 'Name' are different keys
# 'Name' in D
'name' in D
```

```python
# 'name' in D checks only keys not values inside them.

'Mike' in D
```

```python
# To show values inside a Dictionary:

D.values()
```

```python
# To show key/value pairs individually

D.items()
```
