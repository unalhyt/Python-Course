```python
import random # another library to create random numbers

random.randint(1,10) # difference from the np.random(), 10 is included into the random numbers
```

```python
# reminder: single-line if statement

if random.randint(1,10)>5: print('This is a large number')
```

```python
# reminder: single-line if-else statement
print('This is a large number') if random.randint(1,10)<5 else print('This is a small number')
```

```python
for i in range(10):
  print(i**2)
```

```python
# Single-line for loop

[print(i**2) for i in range(10)]
```

```python
# list comprehension

n = [i**2-i**(1/2) for i in range(10) if i>5]
n
```

```python
# convert range to list

newlist = ['']*10

for i in range(10):
  if i%2==1:
    newlist[i] = 'odd'
  else:
    newlist[i] = i

newlist
```

```python
newlist = ['odd' if i%2==1 else i for i in range(10)]
newlist
```
