The variables we assigned above were all numeric variables. If we click on them and wait, we'll see that they're int. This is the first variable type we learned about.

The second variable type we'll learn is **string**. We can create strings with single quotes ' ' or double quotes " ". Python doesn't distinguish between these. However, we can use single quotes as upper quotes inside double-quoted strings.

```python
d = 'Hello'
e = "Mike's favorite color is purple"

print(d, e)
```
