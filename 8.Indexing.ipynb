```python
# indexing : to extract specific items out of a list or an array or a string

arange = range(10,15)
arange
```

```python
alist = list(arange)
alist
```

```python
# python starts counting from zero.

alist[1]

```

```python
alist = [54,63,65,9,0,-644,524]
print(alist[5])
print(alist[-1]) # prints last element
print(alist[-3])
```

```python
# generally for naming a variable for indexing, we use idx

idx = 3
alist[idx]
```

```python
idx = [3,4]
alist[idx]
# this will gives an error message
# to get multiple items from a list, later we will use numpy library
```

```python
alist = [4,3.2,[9,4,6],32]
# question is: how can you get the value 6 (inside the 2. item list)
alist[2][2]

# it works like
# sublist = alist[2]
# sublist[2]
# python makes it easier for us
```

```python
# indexing works also for strings

s = 'hello'

s[0]
```
