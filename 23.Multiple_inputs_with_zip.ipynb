```python
# Multiple inputs with zip

somelist = [3,4,5,6,7]
otherlist = ['d','e','s','m','x']

# want to print d3, e4, s5, m6, x7

for i in range(len(somelist)):
  print(otherlist[i] + ' ' + str(somelist[i]))
```

```python
for j,i in zip(otherlist, somelist):
  print(j, i)
```

```python
z = zip(otherlist, somelist)
z[0]
```

```python
# to reach elements of the z

lz = list(z) # list() converts and destroys the original input
lz
```

```python
lz[1][1]
```

```python
a = [2,3,4,5,6,7,8,4,3,4]
b = ['d','3','f']
c = ['1',2,3,4,6,8]

for i,j,k in zip(a,b,c):
  print(i,j,k)
```

```python
# Exercise: use the zip function to create a dictionary out of two lists.
names  = ['alpha','beta','gamma']
values = [10,20,40]
D = {}
for i,j in zip(names, values):
  D.update({i:j})
print(D)
```

```python
D = dict(zip(names, values))
print(D)
D['gamma']
```
