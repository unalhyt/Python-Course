```python
num = 45.3
sng = 'cellar doors'

print('I would like to have ' + str(num) + ' ' + sng)
```

```python
print('I would like to have %g %s'%(num,sng)) # %g is for number, %s is for string, %i is for integer
print('I would like to have %f %s'%(num,sng)) # %f is for decimal points
print('I would like to have %.3f %s'%(num,sng)) # %.3f is for 3 digits after decimal points
print('I would like to have %10.3f %s'%(num,sng)) # %10.3f creates 10 digits for this number
print('I would like to have %g %s'%(num/5**3,sng)) # you can implement mathematical opearions
```

```python
# f-string
print(f'I would like to have {num} {sng}')
print(f'I would like to have {num/5**3} {sng}')
```

```python
print('I would like to have {n} {s}'.format(n=num,s=sng)) # n and s are created only inside the print function.
```

```python
import string

letters = string.ascii_lowercase
print(letters)
```

```python
for i in range(len(letters)):
    if i == 0 or i == 20:
      ordind = 'st'
    elif i == 1 or i == 21:
      ordind = 'nd'
    elif i == 2 or i == 22:
      ordind = 'rd'
    else:
      ordind = 'th'
    #print('%s is the %g%s letter of the alphabet.'%(letters[i],i+1,ordind))
    print(f'{letters[i]} is the {i+1}{ordind} letter of the alphabet.')
```

```python
for i in range(len(letters)):
    if i%10 == 0 and i != 10:
      ordind = 'st'
    elif i%10 == 1 and i != 11:
      ordind = 'nd'
    elif i%10 == 2 and i != 12:
      ordind = 'rd'
    else:
      ordind = 'th'
    #print('%s is the %g%s letter of the alphabet.'%(letters[i],i+1,ordind))
    print(f'{letters[i]} is the {i+1}{ordind} letter of the alphabet.')
```
