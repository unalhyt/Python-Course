```python
# Slicing

alist = list(range(5,11))
alist
```

```python
# question: how can we access multiple of these elements?

alist[0:2] # we use slicing. 0:2 de second element excluded
```

```python
alist[2:] #from second to the end
```

```python
alist[:4]
```

```python
alist[0:5:2] # Here, 2 is skipping element
```

```python
alist[::3] # from beginning to the end skipping 3
```

```python
alist[::-1] # gives reversed version
```

```python
start = 1
stop  = 4
skip  = 2

alist[start:stop:skip]
```
