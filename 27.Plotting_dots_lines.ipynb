```python
import matplotlib.pyplot as plt

# plotting dots

#plt.plot(3,4)
plt.plot(3,4,'o') # 'o' is the marker
plt.plot(2,4,'s') # plots top of the previous plot
plt.show() # you can use this to differentiate the plots


plt.plot(123,12,'o')
plt.plot(132,14,'s')
plt.show()
```

```python
plt.plot(3,2,'ro') # 'ro' is red circle
plt.plot(3,4,'bp')
plt.show()
```

```python
a = 3

plt.plot(a,a*3/4**2,'ro', label='red circle')
plt.plot(3,4,'bp',label='blue pentagon')
plt.legend() # adds a legend
plt.show()
```

```python
# plotting lines

plt.plot([0,1],[0,2],'r') # you have to specify x and y coordinates of starting and ending points
plt.plot([0,0],[1,2],'g')
plt.legend(['red line', 'green line'])
plt.show()
```

```python
import numpy as np

x = np.linspace(0,3*np.pi,123)
y = np.sin(x)
```

```python
plt.plot(x,y,'k.') # 'k.' for black dots
```
